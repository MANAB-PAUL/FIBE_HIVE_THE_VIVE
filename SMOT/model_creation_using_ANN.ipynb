```python
### reading data set
X_train = pd.read_csv('X_train.csv')
y_train = pd.read_csv('y_train.csv')
X_test = pd.read_csv('X_test.csv')
y_test = pd.read_csv('y_test.csv')
```

```python
### scaling x train
x = X_train.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
X_train_scaled = min_max_scaler.fit_transform(x)
df_X_train_scaled = pd.DataFrame(X_train_scaled)
```

```python
x = X_test.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
X_test_scaled = min_max_scaler.fit_transform(x)
df_X_test_scaled = pd.DataFrame(X_test_scaled)
```

```python
model_ANN = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(19,)),
    layers.Dense(64, activation='relu'),
    layers.Dense(32, activation = 'relu'),
    layers.Dense(2, activation='softmax')
])

# Compile the model
model_ANN.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model_ANN.fit(X_ttrain, y_ttrain, epochs=200)

# Evaluate the model on the test data
loss, accuracy = model_ANN.evaluate(X_ttest, y_ttest)
```

```python
y_pred = model_ANN.predict(X_ttest)
```

```python
yy_pred = []
for i in y_pred:
    if i[0] > i[1]:
        yy_pred.append(0)
    else:
        yy_pred.append(1)
```
